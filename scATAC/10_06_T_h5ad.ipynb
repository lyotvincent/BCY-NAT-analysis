```r
suppressMessages(library(ggplot2))
suppressMessages(library(ArchR))
suppressMessages(library(Seurat))
suppressMessages(library(Signac))
suppressMessages(library(purrr))
suppressMessages(library(readr))
```

```r
set.seed(42)
addArchRThreads(threads = 64)
```

```r
proj <- loadArchRProject("./ArchRProject_T", showLogo = FALSE)
```

```r
## 基因表达矩阵
geneExpr <- getMatrixFromProject(ArchRProj = proj,
                             useMatrix = "GeneScoreMatrix")
```

```r
cell_annotations <- proj@cellColData
cell_annotations$predictedGroup
```

```r
gene_names <- rowData(geneExpr)$name
```

```r
lsiMatrix <- getReducedDims(ArchRProj = proj, reducedDims = "IterativeLSI", corCutOff = 1)
```

```r
lsiMatrix
```

```r
# 导出峰值矩阵
writeMM(geneExpr@assays@data$GeneScoreMatrix, "../output/joint_annotation/T_geneScore_matrix.mtx")
write.csv(as.data.frame(lsiMatrix), "../output/joint_annotation/T_lsiMatrix.csv", row.names = TRUE)
write.csv(as.data.frame(cell_annotations), "../output/joint_annotation/T_cell_annotations.csv")
```

```r
write.csv(gene_names, "../output/joint_annotation/T_gene_names.csv", row.names = FALSE, col.names = FALSE)
```
