```r
set.seed(42)
addArchRThreads(threads = 64)
```

```r
proj <- loadArchRProject("./ArchRProject_T", showLogo = FALSE)
```

```r
batches <- c("b2g_T", 'b2g_T_batch', 'b2g_T_raw')
```

```r
# 为每个批次生成文件名
save_dataPath = "/data/xwt/BCY_ATAC/B2G/T/"
files <- paste0(save_dataPath, batches, ".csv")
files 
```

```r
df1 = read_csv(file = files[1])
colnames(df1)[34] <- "leiden_T"
```

```r
df2 = read_csv(file = files[2])
colnames(df2)[34] <- "leiden_T"
```

```r
df3 = read_csv(file = files[3])
colnames(df3)[34] <- "leiden_T"
```

```r
proj@cellColData$leiden_T <- df1$leiden_T[match(rownames(proj@cellColData), df1$cell_id)]
```

```r
proj@cellColData$leiden_T_batch <- df2$leiden_T[match(rownames(proj@cellColData), df2$cell_id)]
```

```r
proj@cellColData$leiden_T_raw <- df3$leiden_T[match(rownames(proj@cellColData), df3$cell_id)]
```

```r
proj$leiden_T
```

```r
### motif enrichment
VarMotifs <- getVarDeviations(proj, name = "homerMatrix", plot = F)
MotifScoreClusters <- getMarkerFeatures(proj, 
                                        useMatrix = "homerMatrix", 
                                        groupBy = "leiden_T",
                                        bias = c("TSSEnrichment", "log10(nFrags)"),
                                        testMethod = "wilcoxon", 
                                        useSeqnames = "z")

fh <- function(x) hclust(dist(x), method="ward.D2")
idy <- which(mcols(MotifScoreClusters)$name %in% VarMotifs$name[c(1:50)])
mtx2 <- assays(MotifScoreClusters[idy,])$Mean %>% `rownames<-`(., mcols(MotifScoreClusters)$name[idy])

col_fun2 <- colorRamp2(c(-5,-2.5,0,2,5), paletteContinuous(set = "solarExtra", n = 5))
ht2 <- Heatmap(mtx2, name = "Mean motif score", cluster_rows = fh, cluster_columns = F, show_row_dend = F, col = col_fun2)
svg("/data/hanxue/BCY_ATAC/joint_output/Motif-leiden-T.svg",width = 10, height = 10)
options(repr.plot.width=10, repr.plot.height=10)
p6 <- draw(ht2)
dev.off()
```
