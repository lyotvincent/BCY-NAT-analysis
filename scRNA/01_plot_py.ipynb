```python
import os,sys
%matplotlib inline
import matplotlib.pylab as plt
import numpy as np
plt.rcParams['figure.dpi'] = 300
plt.rcParams['savefig.dpi']=600
plt.rcParams['font.family']='sans serif' #please remove this line if font is not installed
plt.rcParams['font.sans-serif']='Arial' # please remove this line if Arial font is not installed
plt.rcParams['pdf.fonttype']=42
import PyComplexHeatmap as pch
import pandas as pd
```

```python
data = {
    'YC02': ['0', '0', '1', '0', '0', '0', '0', '0', 'PR', 'TNBC'],  # 1 
    'YD45': ['0', '0', '1', '0', '1', '0', '1', '0', 'SD', 'TNBC'], # 2
    'YZ01': ['0', '0', '1', '0', '0', '0', '0', '0', 'PR', 'TNBC'], # 8
    'YX92': ['0', '0', '1', '0', '0', '0', '0', '0', 'PCR', 'TNBC'], # 9
    'YZ20': ['1', '1', '0', '0', '0', '0', '1', '1', 'None', 'Luminal B'], # 10 
    'YL03': ['0', '0', '0', '0', '1', '1', '0', '0', 'PR', 'Luminal B'],  # 11
    'YK37': ['0', '0', '1', '1', '0', '0', '1', '1', 'SD', 'Luminal B'], # 12
    'YC82': ['0', '0', '0', '0', '1', '1', '0', '0', 'PR', 'Luminal B'], # 13
    'YS05': ['0', '0', '1', '0', '0', '0', '0', '0', 'SD', 'Luminal B'], # 15
    'YY65': ['0', '0', '1', '0', '0', '0', '0', '0', 'PR', 'HR-HER2+'],  # 17
    'YL37': ['0', '1', '0', '0', '0', '0', '0', '0', 'None', 'HR+HER2-'], # 18
    'YY60': ['0', '0', '0', '0', '1', '1', '0', '0', 'SD', 'HR+HER2+'],    # 19
    'YW10': ['1', '1', '0', '0', '0', '0', '0', '0', 'None', 'HR+HER2+'], # 20
    'YY12': ['1', '1', '0', '0', '0', '0', '0', '0', 'None', 'HR+HER2+'], # 21
    'YG43': ['1', '1', '0', '0', '0', '0', '0', '0', 'None', 'HR+HER2+'] # 22
    }
df = pd.DataFrame(data, index=[
    'Non-NAT scRNA-seq', 
    'Non-NAT scATAC-seq', 
    'Pre-NAT scRNA-seq', 
    'Pre-NAT scATAC-seq', 
    'Post-NAT scRNA-seq', 
    'Post-NAT scATAC-seq', 
    'Non-Tumor scRNA-seq', 
    'Non-Tumor scATAC-seq', 
    'NAR', 
    'Subtype'])

df = df
df
df_heatmap = df.iloc[:8, :].astype(float)
# df_heatmap
df_info = df.iloc[8:, :].T
# df_info
```

```python
row_ha = pch.HeatmapAnnotation(Subtype=pch.anno_simple(df_info.Subtype, cmap='Set1', height=5,legend=True), 
                               NAR=pch.anno_simple(df_info.NAR, cmap='Pastel1', height=5,legend=True), 
                               axis=0,legend=True,legend_gap=5,wgap=2,
                               label_kws=dict(visible=True,color='black',rotation=30,fontsize=5,verticalalignment='center',horizontalalignment='left'))
plt.figure(figsize=(3, 5))
cm = pch.ClusterMapPlotter(data=df_heatmap.T,right_annotation=row_ha,legend_hpad=2,label='values',
                       show_rownames=True,show_colnames=True,row_cluster=False, col_cluster=False,
                       row_names_side='left',verbose=0,
                       col_names_side='top',
                       cmap='GnBu',xticklabels_kws={'labelrotation':30},linewidths=0.05,linecolor='gold',yticklabels_kws={'pad':1},
                    #    xlabel_kws = dict(color='white',fontsize=4,labelpad=25),
                       subplot_gap=2)
plt.savefig('./figures/Sample_Info.svg', format='svg', bbox_inches='tight')
plt.show()
```

```python
col_ha = pch.HeatmapAnnotation(Subtype=pch.anno_simple(df_info.Subtype, cmap='Set1', height=3,legend=True), 
                               NAR=pch.anno_simple(df_info.NAR, cmap='Pastel1', height=3,legend=True), 
                               axis=1,legend=True,legend_gap=5,hgap=1)
plt.figure(figsize=(5, 2.5))
cm = pch.ClusterMapPlotter(data=df_heatmap,top_annotation=col_ha,legend_hpad=2,label='values',
                       show_rownames=True,show_colnames=True,row_cluster=False, col_cluster=False,
                       row_names_side='right',verbose=0,
                       cmap='GnBu',xticklabels_kws={'labelrotation':45},linewidths=0.05,linecolor='gold',yticklabels_kws={'pad':1})
plt.savefig('./figures/Sample_Info_2.svg', format='svg', bbox_inches='tight')
plt.show()
```
